# overfitting

```python
x=np.random.rand(6)
y=np.random.rand(6)
```

```python
plt.scatter(x,y,c='r',s=50)
```

```python
def myplot(n=1):
    plt.scatter(x,y,c='r',s=50)
    t=np.linspace(-0.2,1.2,1000)
    plt.plot(t,0.5*np.sin(n*t)+0.5)
```

```python
myplot(200)
```

```python
from ipywidgets import interact
```

```python
interact(myplot,n=(1,200))
```
